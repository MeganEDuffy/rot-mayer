### Evaluating and visualizing NAAF-adjusted bacterial peptide composition and GO terms during rot experiment

#### Beginning with: bacterial LCA assigments adjusted by NAAF factor in csv, manually created from bacterial LCA NAAF csvs

#### Goal: heatmaps and stacked bar plot bacterial peptide contribution over the 4 timepoints

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
from functools import reduce
```

```python
# rearranged and condensed the csv manually in Libre Calc
# PERCENT of all NAAF terms
# read in new csv:

bac_PER = pd.read_csv("/home/millieginty/Documents/git-repos/rot-mayer/analyses/unipept/lca/bacterial-comp-NAAF/Rot-bacterial-comp_DN_percent_sorted.csv", index_col=0)

bac_PER.head()
```

```python
bac_TPER = bac_PER.transpose()

ax = bac_TPER.plot(kind='bar',stacked=True, figsize = (5,7))

ax.set_xlabel("Rot progression", fontsize = 16)

ax.set_ylabel("Relative peptide abundance", labelpad=10, fontsize = 16)

labels = ['Day 0', 'Day 2', 'Day 5', 'Day 12']

ax.set_xticklabels(labels, rotation = 45, size = 14)
plt.yticks(fontsize=14)

# For some reason the order of the handles command (in front of the legend loc command matters here)
handles, labels = ax.get_legend_handles_labels()
ax.legend(handles[::-1], labels[::-1])

plt.legend(bbox_to_anchor=(0, -0.74, 1, 0), loc="lower left", mode="expand", ncol=1)


plt.title("A. Bacterial class-level peptide identifications", size = 16, pad = 20)

plt.savefig('/home/millieginty/Documents/git-repos/rot-mayer/figures/manuscript-files/fig6a-bac.png', bbox_inches='tight', dpi=400)
```

```python
# rearranged and condensed the csv manually in Libre Calc
# read in new csv:

bac = pd.read_csv("/home/millieginty/Documents/git-repos/rot-mayer/analyses/unipept/GO-terms/DN/TD+ND/DN80_combined_TD+ND_bp_GO_sorted_combined.csv", index_col=0)

bac.head()
```

```python
sns.set(font_scale=1.2)  

fig, ax = plt.subplots(figsize=(6,12))

ax = sns.heatmap(bac, cmap="BuPu", annot=True, linecolor = 'grey', linewidths=.01, \
                square=False, cbar_kws={'label': '% of all bacterial GO terms'}, label='big')

ax.set_ylabel('')

# turn the axis label
for item in ax.get_yticklabels():
    item.set_rotation(0)
    

for item in ax.get_xticklabels():
    item.set_rotation(45)
    
plt.title("B. Bacterial biological process GO terms", size = 16, pad = 20)

plt.savefig('/home/millieginty/Documents/git-repos/rot-mayer/figures/manuscript-files/fig6b-bacgo.png', bbox_inches='tight', dpi=400)
```
