### Manipulation of PeaskDB de novo-assisted database search results of _T. weisflogii_ rot samples LC-MS/MS data using python.

Starting with:

    PeaksDB search results (.csv) of PTM-optimized database searches
    These were all searched with 15 ppm precursor tolerance and 0.5 ppm fragement ion tolerance
    Multiple injections and fragmentation strategies included
    Exported at <1.0% FDR
    
Goal:

    Files with stripped (no PTMs) peptide lists and
    Columns with #'s of each modification in every sequence
    Column with stripped peptide lengths (# amino acids)
    
### To use for a different file:

#### 1. Change the input file name in *IN 4*
#### 2. Use 'find + replace' (Esc + F) to replace the running # (e.g., 233) for another
#### 3. Update the NAAF factor calculated in *IN 6* into *IN 7*

```python
# read the CSV into a dataframe using the pandas read_csv function
peaksdbdup332 = pd.read_csv("/home/millieginty/Documents/git-repos/rot-mayer/data/MED_Weissrot_Fusion_UWPR2021/MED_Weissrot_Fusion_332-T12nd-all_PEAKS_101/332-T12nd-all_PEAKS_101_peptide.csv")

# remove redundant rows
peaksdb332 = pd.DataFrame.drop_duplicates(peaksdbdup332)

print(peaksdb332.columns)

columns = ['Peptide', '-10lgP', 'Mass', 'Length', 'ppm', 'm/z', 'RT',
       'Area', 'Fraction', 'Scan', 'Source File',
       '#Spec', '#Spec', 'Accession', 'PTM',
       'AScore']

peaksdb332.columns = columns

print("# redundant peaksdb peptides in combined dataframe", len(peaksdbdup332))
print("# nonredundant peaksdb peptides in combined dataframe", len(peaksdb332))

mean_length = peaksdb332['Length'].mean()
print(mean_length)

#look at the dataframe
peaksdb332.head()
```

```python
# use a count function to enumerate the # of A's (alanines) in each peptide
peaksdb332['A'] = peaksdb332['Peptide'].str.count("A")

# use a count function to enumerate the # of C's (cysteines) in each peptide
peaksdb332['C'] = peaksdb332['Peptide'].str.count("C")

# use a count function to enumerate the # of D's (aspartic acids) in each peptide
peaksdb332['D'] = peaksdb332['Peptide'].str.count("D")

# use a count function to enumerate the # of E's (glutamic acids) in each peptide
peaksdb332['E'] = peaksdb332['Peptide'].str.count("E")

# use a count function to enumerate the # of F's (phenylalanines) in each peptide
peaksdb332['F'] = peaksdb332['Peptide'].str.count("F")

# use a count function to enumerate the # of G's (glycines) in each peptide
peaksdb332['G'] = peaksdb332['Peptide'].str.count("G")

# use a count function to enumerate the # of H's (histidines) in each peptide
peaksdb332['H'] = peaksdb332['Peptide'].str.count("H")

# use a count function to enumerate the # of I's (isoleucines) in each peptide
# in peaksdb332 output, there will be no isoleucines (they're lumped in with leucines)
peaksdb332['I'] = peaksdb332['Peptide'].str.count("I")

# use a count function to enumerate the # of K's (lysines) in each peptide
peaksdb332['K'] = peaksdb332['Peptide'].str.count("K")

# use a count function to enumerate the # of L's (leucines) in each peptide
# also these include the isoleucines
peaksdb332['L'] = peaksdb332['Peptide'].str.count("L")

# use a count function to enumerate the # of M's (methionines) in each peptide
peaksdb332['M'] = peaksdb332['Peptide'].str.count("M")

# use a count function to enumerate the # of N's (asparagines) in each peptide
peaksdb332['N'] = peaksdb332['Peptide'].str.count("N")

# use a count function to enumerate the # of P's ([prolines]) in each peptide
peaksdb332['P'] = peaksdb332['Peptide'].str.count("P")

# use a count function to enumerate the # of Q's (glutamines) in each peptide
peaksdb332['Q'] = peaksdb332['Peptide'].str.count("Q")

# use a count function to enumerate the # of R's (arginines) in each peptide
peaksdb332['R'] = peaksdb332['Peptide'].str.count("R")

# use a count function to enumerate the # of S's (serines) in each peptide
peaksdb332['S'] = peaksdb332['Peptide'].str.count("S")

# use a count function to enumerate the # of T's (threonines) in each peptide
peaksdb332['T'] = peaksdb332['Peptide'].str.count("T")

# use a count function to enumerate the # of V's (valines) in each peptide
peaksdb332['V'] = peaksdb332['Peptide'].str.count("V")

# use a count function to enumerate the # of W's (tryptophans) in each peptide
peaksdb332['W'] = peaksdb332['Peptide'].str.count("W")

# use a count function to enumerate the # of Y's (tyrosines) in each peptide
peaksdb332['Y'] = peaksdb332['Peptide'].str.count("Y")

# use a count function to enumerate the # of carbamidomethylated C's in each peptide
peaksdb332['c-carb'] = peaksdb332['Peptide'].str.count("57.02")

# use a count function to enumerate the # of oxidized M's in each peptide
peaksdb332['m-oxid'] = peaksdb332['Peptide'].apply(lambda x: x.count('M(+15.99)'))

# use a count function to enumerate the # of oxidized K's in each peptide
peaksdb332['k-oxid'] = peaksdb332['Peptide'].apply(lambda x: x.count('K(+15.99)'))

# use a count function to enumerate the # of oxidized P's in each peptide
peaksdb332['p-oxid'] = peaksdb332['Peptide'].apply(lambda x: x.count('P(+15.99)'))

# use a count function to enumerate the # of oxidized R's in each peptide
peaksdb332['r-oxid'] = peaksdb332['Peptide'].apply(lambda x: x.count('R(+15.99)'))

# use a count function to enumerate the # of oxidized Y's in each peptide
peaksdb332['y-oxid'] = peaksdb332['Peptide'].apply(lambda x: x.count('Y(+15.99)'))

# use a lamba function to enumerate the # of deamidated N's in each peptide
peaksdb332['n-deam'] = peaksdb332['Peptide'].apply(lambda x: x.count('N(+.98)'))

# use a count function to enumerate the # of methylated K's in each peptide
peaksdb332['k-meth'] = peaksdb332['Peptide'].apply(lambda x: x.count('K(+14.02)'))

# use a count function to enumerate the # of methylated R's in each peptide
peaksdb332['r-meth'] = peaksdb332['Peptide'].apply(lambda x: x.count('R(+14.02)'))

# use a count function to enumerate the # of pyro glu Q's in each peptide
peaksdb332['q-pyro'] = peaksdb332['Peptide'].apply(lambda x: x.count('Q(-17.03)'))

# use a count function to enumerate the # of acetylation of K's in each peptide
peaksdb332['k-acet'] = peaksdb332['Peptide'].apply(lambda x: x.count('K(+42.01)'))

# create a column with 'stripped' peptide sequences using strip
peaksdb332['stripped peptide'] = peaksdb332['Peptide'].str.replace(r"\(.*\)","")

# add a column with the stripped peptide length (number of AAs)
peaksdb332['stripped length'] = peaksdb332['stripped peptide'].apply(len)

peaksdb332['NAAF num.'] = peaksdb332['Area'] / peaksdb332['stripped length']

# total the number of modifications in sequence
peaksdb332['ptm-total'] = peaksdb332['c-carb'] + peaksdb332['m-oxid'] + peaksdb332['k-oxid'] + peaksdb332['p-oxid'] \
+ peaksdb332['r-oxid'] + peaksdb332['y-oxid'] + peaksdb332['n-deam'] + peaksdb332['k-meth'] + peaksdb332['r-meth'] \
+ peaksdb332['q-pyro'] + peaksdb332['k-acet']

# turn all isoleucines into leucines
# this helps later in comparing Unipept peptides to PeaksDB and Comet ones
peaksdb332['stripped I-L']= peaksdb332['stripped peptide'].str.replace('I','L')

# write modified dataframe to new txt file
peaksdb332.to_csv("/home/millieginty/Documents/git-repos/rot-mayer/data/processed/PeaksDB/TW_332_T12_undigested_combine_PTMopt_DB_FDR1.csv")

# check out the results
peaksdb332.head()
```

## NAAF correction and exporting files with AA and PTM totals:

```python
# made a new dataframe that contains the sums of certain columns in the modified
# peptide dataframe above 

index = ['sample total']

data = {'A': peaksdb332['A'].sum(),
        'C': peaksdb332['C'].sum(),
        'D': peaksdb332['D'].sum(),
        'E': peaksdb332['E'].sum(),
        'F': peaksdb332['F'].sum(),
        'G': peaksdb332['G'].sum(),
        'H': peaksdb332['H'].sum(),
        'I': peaksdb332['I'].sum(),
        'K': peaksdb332['K'].sum(),
        'L': peaksdb332['L'].sum(),
        'M': peaksdb332['M'].sum(),
        'N': peaksdb332['N'].sum(),
        'P': peaksdb332['P'].sum(),
        'Q': peaksdb332['Q'].sum(),
        'R': peaksdb332['R'].sum(),
        'S': peaksdb332['S'].sum(),
        'T': peaksdb332['T'].sum(),
        'V': peaksdb332['V'].sum(),
        'W': peaksdb332['W'].sum(),
        'Y': peaksdb332['Y'].sum(),
        'c-carb': peaksdb332['c-carb'].sum(),
        'm-oxid': peaksdb332['m-oxid'].sum(),
        'k-oxid': peaksdb332['k-oxid'].sum(),
        'p-oxid': peaksdb332['p-oxid'].sum(),
        'r-oxid': peaksdb332['r-oxid'].sum(),
        'y-oxid': peaksdb332['y-oxid'].sum(),
        'n-deam': peaksdb332['n-deam'].sum(),
        'k-meth': peaksdb332['k-meth'].sum(),
        'r-meth': peaksdb332['r-meth'].sum(),
        'q-pyro': peaksdb332['q-pyro'].sum(),
        'k-acet': peaksdb332['k-acet'].sum(),
        'Total area': peaksdb332['Area'].sum(),
        'Total length': peaksdb332['stripped length'].sum()
       }

totalpeaksdb332 = pd.DataFrame(data, columns=['A', 'C', 'D', 'E', 'F', 'G', 'H', 'I', 'K', 'L', \
                                              'M', 'N', 'P', 'Q', 'R', 'S', 'T', 'V', 'W', 'Y', \
                                              'c-carb', 'm-oxid', 'k-oxid', 'p-oxid', 'r-oxid', \
                                              'y-oxid', 'n-deam', 'k-meth', 'r-meth', 'q-pyro', \
                                              'k-acet', 'Total area', 'Total length'], index=index)

# calculate percentage of C's with carb (should be 1.0)
totalpeaksdb332['% C w/ carb'] = totalpeaksdb332['c-carb'] / totalpeaksdb332['C'] 

# calculate percentage of M's that are oxidized
totalpeaksdb332['% M w/ oxid'] = totalpeaksdb332['m-oxid'] / totalpeaksdb332['M'] 

# calculate percentage of K's that are oxidized
totalpeaksdb332['% K w/ oxid'] = totalpeaksdb332['k-oxid'] / totalpeaksdb332['K'] 

# calculate percentage of P's that are oxidized
totalpeaksdb332['% P w/ oxid'] = totalpeaksdb332['p-oxid'] / totalpeaksdb332['P'] 

# calculate percentage of R's that are oxidized
totalpeaksdb332['% R w/ oxid'] = totalpeaksdb332['p-oxid'] / totalpeaksdb332['R'] 

# calculate percentage of Y's that are oxidized
totalpeaksdb332['% Y w/ oxid'] = totalpeaksdb332['y-oxid'] / totalpeaksdb332['Y'] 

# calculate percentage of N's that are deamidated
totalpeaksdb332['% N w/ deam'] = totalpeaksdb332['n-deam'] / totalpeaksdb332['N'] 

# calculate percentage of K's that are methylated
totalpeaksdb332['% K w/ meth'] = totalpeaksdb332['k-meth'] / totalpeaksdb332['K'] 

# calculate percentage of R's that are methylated
totalpeaksdb332['% R w/ meth'] = totalpeaksdb332['r-meth'] / totalpeaksdb332['R'] 

# calculate percentage of Q's that are pyro glu'd
totalpeaksdb332['% Q w/ pyro'] = totalpeaksdb332['q-pyro'] / totalpeaksdb332['Q'] 

# calculate percentage of K's that are acetylation
totalpeaksdb332['% K w/ acet'] = totalpeaksdb332['k-acet'] / totalpeaksdb332['K'] 

# calculate NAAF denominator for all peptides in dataset i
totalpeaksdb332['NAAF denom.'] = totalpeaksdb332['Total area'] / totalpeaksdb332['Total length']

# write modified dataframe to new txt file
totalpeaksdb332.to_csv("/home/millieginty/Documents/git-repos/rot-mayer/data/processed/PeaksDB/TW_332_T12_undigested_combine_PTMopt_DB_FDR1_totals.csv")

totalpeaksdb332.head()
```

```python
# use the calculated NAAF factor (in totalpeaksdb332 dataframe, above) to caluclate the NAAF 
# NAAF: normalized normalized area abundance factor

NAAF20 = 1.491534e+06

# we'll use -10lgP > 20 as an approximate cutoff for the <1.0% FDR which is what we actually have
peaksdb332['NAAF factor'] = (peaksdb332['NAAF num.'])/NAAF20

# make a dataframe that contains only what we need: sequences, AAs, PTMs
peaksdb332_NAAF = peaksdb332[['stripped peptide', 'NAAF factor', 'A', 'C', 'D', 'E', 'F', 'G', 'H', \
                              'I', 'K', 'L', 'M', 'N', 'P', 'Q', 'R', 'S', 'T', 'V', 'W', 'Y', \
                              'c-carb', 'm-oxid', 'k-oxid', 'p-oxid', 'r-oxid', 'y-oxid', 'n-deam', \
                              'k-meth', 'r-meth', 'q-pyro', 'k-acet']].copy()

# multiply the NAAF20 factor by the AAs to normalize its abundance by peak area and peptide length

peaksdb332_NAAF['A-NAAF20'] = peaksdb332_NAAF['A'] * peaksdb332['NAAF factor']
peaksdb332_NAAF['C-NAAF20'] = peaksdb332_NAAF['C'] * peaksdb332['NAAF factor']
peaksdb332_NAAF['D-NAAF20'] = peaksdb332_NAAF['D'] * peaksdb332['NAAF factor']
peaksdb332_NAAF['E-NAAF20'] = peaksdb332_NAAF['E'] * peaksdb332['NAAF factor']
peaksdb332_NAAF['F-NAAF20'] = peaksdb332_NAAF['F'] * peaksdb332['NAAF factor']
peaksdb332_NAAF['G-NAAF20'] = peaksdb332_NAAF['G'] * peaksdb332['NAAF factor']
peaksdb332_NAAF['H-NAAF20'] = peaksdb332_NAAF['H'] * peaksdb332['NAAF factor']
peaksdb332_NAAF['I-NAAF20'] = peaksdb332_NAAF['I'] * peaksdb332['NAAF factor']
peaksdb332_NAAF['K-NAAF20'] = peaksdb332_NAAF['K'] * peaksdb332['NAAF factor']
peaksdb332_NAAF['L-NAAF20'] = peaksdb332_NAAF['L'] * peaksdb332['NAAF factor']
peaksdb332_NAAF['M-NAAF20'] = peaksdb332_NAAF['M'] * peaksdb332['NAAF factor']
peaksdb332_NAAF['N-NAAF20'] = peaksdb332_NAAF['N'] * peaksdb332['NAAF factor']
peaksdb332_NAAF['P-NAAF20'] = peaksdb332_NAAF['P'] * peaksdb332['NAAF factor']
peaksdb332_NAAF['Q-NAAF20'] = peaksdb332_NAAF['Q'] * peaksdb332['NAAF factor']
peaksdb332_NAAF['R-NAAF20'] = peaksdb332_NAAF['R'] * peaksdb332['NAAF factor']
peaksdb332_NAAF['S-NAAF20'] = peaksdb332_NAAF['S'] * peaksdb332['NAAF factor']
peaksdb332_NAAF['T-NAAF20'] = peaksdb332_NAAF['T'] * peaksdb332['NAAF factor']
peaksdb332_NAAF['V-NAAF20'] = peaksdb332_NAAF['V'] * peaksdb332['NAAF factor']
peaksdb332_NAAF['W-NAAF20'] = peaksdb332_NAAF['W'] * peaksdb332['NAAF factor']
peaksdb332_NAAF['Y-NAAF20'] = peaksdb332_NAAF['Y'] * peaksdb332['NAAF factor']

# multiply the NAAF20 factor by the PTMs normalize its abundance by peak area and peptide length

peaksdb332_NAAF['ccarb-NAAF20'] = peaksdb332_NAAF['c-carb'] * peaksdb332_NAAF['NAAF factor']
peaksdb332_NAAF['moxid-NAAF20'] = peaksdb332_NAAF['m-oxid'] * peaksdb332_NAAF['NAAF factor']
peaksdb332_NAAF['koxid-NAAF20'] = peaksdb332_NAAF['k-oxid'] * peaksdb332_NAAF['NAAF factor']
peaksdb332_NAAF['poxid-NAAF20'] = peaksdb332_NAAF['p-oxid'] * peaksdb332_NAAF['NAAF factor']
peaksdb332_NAAF['roxid-NAAF20'] = peaksdb332_NAAF['r-oxid'] * peaksdb332_NAAF['NAAF factor']
peaksdb332_NAAF['yoxid-NAAF20'] = peaksdb332_NAAF['y-oxid'] * peaksdb332_NAAF['NAAF factor']
peaksdb332_NAAF['ndeam-NAAF20'] = peaksdb332_NAAF['n-deam'] * peaksdb332_NAAF['NAAF factor']
peaksdb332_NAAF['kmeth-NAAF20'] = peaksdb332_NAAF['k-meth'] * peaksdb332_NAAF['NAAF factor']
peaksdb332_NAAF['rmeth-NAAF20'] = peaksdb332_NAAF['r-meth'] * peaksdb332_NAAF['NAAF factor']
peaksdb332_NAAF['qpyro-NAAF20'] = peaksdb332_NAAF['q-pyro'] * peaksdb332_NAAF['NAAF factor']
peaksdb332_NAAF['kacet-NAAF20'] = peaksdb332_NAAF['k-acet'] * peaksdb332_NAAF['NAAF factor']

# write the dataframe to a new csv
peaksdb332_NAAF.to_csv("/home/millieginty/Documents/git-repos/rot-mayer/data/processed/PeaksDB/TW_332_T12_undigested_combine_PTMopt_DB_FDR1_NAAF.csv")

peaksdb332_NAAF.head()
```
