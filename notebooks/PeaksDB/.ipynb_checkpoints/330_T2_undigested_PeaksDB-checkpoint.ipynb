### Manipulation of PeaskDB de novo-assisted database search results of _T. weisflogii_ rot samples LC-MS/MS data using python.

Starting with:

    PeaksDB search results (.csv) of PTM-optimized database searches
    These were all searched with 15 ppm precursor tolerance and 0.5 ppm fragement ion tolerance
    Multiple injections and fragmentation strategies included
    Exported at <1.0% FDR
    
Goal:

    Files with stripped (no PTMs) peptide lists and
    Columns with #'s of each modification in every sequence
    Column with stripped peptide lengths (# amino acids)
    
### To use for a different file:

#### 1. Change the input file name in *IN 4*
#### 2. Use 'find + replace' (Esc + F) to replace the running # (e.g., 233) for another
#### 3. Update the NAAF factor calculated in *IN 6* into *IN 7*

```python
# read the CSV into a dataframe using the pandas read_csv function
peaksdbdup330 = pd.read_csv("/home/millieginty/Documents/git-repos/rot-mayer/data/MED_Weissrot_Fusion_UWPR2021/MED_Weissrot_Fusion_330-T2nd-all_PEAKS_94/330-T2nd-all_PEAKS_94_peptide.csv")

# remove redundant rows
peaksdb330 = pd.DataFrame.drop_duplicates(peaksdbdup330)

print(peaksdb330.columns)

columns = ['Peptide', '-10lgP', 'Mass', 'Length', 'ppm', 'm/z', 'RT',
       'Area', 'Fraction', 'Scan', 'Source File',
       '#Spec', '#Spec', 'Accession', 'PTM',
       'AScore']

peaksdb330.columns = columns

print("# redundant peaksdb peptides in combined dataframe", len(peaksdbdup330))
print("# nonredundant peaksdb peptides in combined dataframe", len(peaksdb330))

mean_length = peaksdb330['Length'].mean()
print(mean_length)

#look at the dataframe
peaksdb330.head()
```

```python
# use a count function to enumerate the # of A's (alanines) in each peptide
peaksdb330['A'] = peaksdb330['Peptide'].str.count("A")

# use a count function to enumerate the # of C's (cysteines) in each peptide
peaksdb330['C'] = peaksdb330['Peptide'].str.count("C")

# use a count function to enumerate the # of D's (aspartic acids) in each peptide
peaksdb330['D'] = peaksdb330['Peptide'].str.count("D")

# use a count function to enumerate the # of E's (glutamic acids) in each peptide
peaksdb330['E'] = peaksdb330['Peptide'].str.count("E")

# use a count function to enumerate the # of F's (phenylalanines) in each peptide
peaksdb330['F'] = peaksdb330['Peptide'].str.count("F")

# use a count function to enumerate the # of G's (glycines) in each peptide
peaksdb330['G'] = peaksdb330['Peptide'].str.count("G")

# use a count function to enumerate the # of H's (histidines) in each peptide
peaksdb330['H'] = peaksdb330['Peptide'].str.count("H")

# use a count function to enumerate the # of I's (isoleucines) in each peptide
# in peaksdb330 output, there will be no isoleucines (they're lumped in with leucines)
peaksdb330['I'] = peaksdb330['Peptide'].str.count("I")

# use a count function to enumerate the # of K's (lysines) in each peptide
peaksdb330['K'] = peaksdb330['Peptide'].str.count("K")

# use a count function to enumerate the # of L's (leucines) in each peptide
# also these include the isoleucines
peaksdb330['L'] = peaksdb330['Peptide'].str.count("L")

# use a count function to enumerate the # of M's (methionines) in each peptide
peaksdb330['M'] = peaksdb330['Peptide'].str.count("M")

# use a count function to enumerate the # of N's (asparagines) in each peptide
peaksdb330['N'] = peaksdb330['Peptide'].str.count("N")

# use a count function to enumerate the # of P's ([prolines]) in each peptide
peaksdb330['P'] = peaksdb330['Peptide'].str.count("P")

# use a count function to enumerate the # of Q's (glutamines) in each peptide
peaksdb330['Q'] = peaksdb330['Peptide'].str.count("Q")

# use a count function to enumerate the # of R's (arginines) in each peptide
peaksdb330['R'] = peaksdb330['Peptide'].str.count("R")

# use a count function to enumerate the # of S's (serines) in each peptide
peaksdb330['S'] = peaksdb330['Peptide'].str.count("S")

# use a count function to enumerate the # of T's (threonines) in each peptide
peaksdb330['T'] = peaksdb330['Peptide'].str.count("T")

# use a count function to enumerate the # of V's (valines) in each peptide
peaksdb330['V'] = peaksdb330['Peptide'].str.count("V")

# use a count function to enumerate the # of W's (tryptophans) in each peptide
peaksdb330['W'] = peaksdb330['Peptide'].str.count("W")

# use a count function to enumerate the # of Y's (tyrosines) in each peptide
peaksdb330['Y'] = peaksdb330['Peptide'].str.count("Y")

# use a count function to enumerate the # of carbamidomethylated C's in each peptide
peaksdb330['c-carb'] = peaksdb330['Peptide'].str.count("57.02")

# use a count function to enumerate the # of oxidized M's in each peptide
peaksdb330['m-oxid'] = peaksdb330['Peptide'].apply(lambda x: x.count('M(+15.99)'))

# use a count function to enumerate the # of oxidized K's in each peptide
peaksdb330['k-oxid'] = peaksdb330['Peptide'].apply(lambda x: x.count('K(+15.99)'))

# use a count function to enumerate the # of oxidized P's in each peptide
peaksdb330['p-oxid'] = peaksdb330['Peptide'].apply(lambda x: x.count('P(+15.99)'))

# use a count function to enumerate the # of oxidized R's in each peptide
peaksdb330['r-oxid'] = peaksdb330['Peptide'].apply(lambda x: x.count('R(+15.99)'))

# use a count function to enumerate the # of oxidized Y's in each peptide
peaksdb330['y-oxid'] = peaksdb330['Peptide'].apply(lambda x: x.count('Y(+15.99)'))

# use a lamba function to enumerate the # of deamidated N's in each peptide
peaksdb330['n-deam'] = peaksdb330['Peptide'].apply(lambda x: x.count('N(+.98)'))

# use a count function to enumerate the # of methylated K's in each peptide
peaksdb330['k-meth'] = peaksdb330['Peptide'].apply(lambda x: x.count('K(+14.02)'))

# use a count function to enumerate the # of methylated R's in each peptide
peaksdb330['r-meth'] = peaksdb330['Peptide'].apply(lambda x: x.count('R(+14.02)'))

# use a count function to enumerate the # of pyro glu Q's in each peptide
peaksdb330['q-pyro'] = peaksdb330['Peptide'].apply(lambda x: x.count('Q(-17.03)'))

# use a count function to enumerate the # of acetylation of K's in each peptide
peaksdb330['k-acet'] = peaksdb330['Peptide'].apply(lambda x: x.count('K(+42.01)'))

# create a column with 'stripped' peptide sequences using strip
peaksdb330['stripped peptide'] = peaksdb330['Peptide'].str.replace(r"\(.*\)","")

# add a column with the stripped peptide length (number of AAs)
peaksdb330['stripped length'] = peaksdb330['stripped peptide'].apply(len)

peaksdb330['NAAF num.'] = peaksdb330['Area'] / peaksdb330['stripped length']

# total the number of modifications in sequence
peaksdb330['ptm-total'] = peaksdb330['c-carb'] + peaksdb330['m-oxid'] + peaksdb330['k-oxid'] + peaksdb330['p-oxid'] \
+ peaksdb330['r-oxid'] + peaksdb330['y-oxid'] + peaksdb330['n-deam'] + peaksdb330['k-meth'] + peaksdb330['r-meth'] \
+ peaksdb330['q-pyro'] + peaksdb330['k-acet']

# turn all isoleucines into leucines
# this helps later in comparing Unipept peptides to PeaksDB and Comet ones
peaksdb330['stripped I-L']= peaksdb330['stripped peptide'].str.replace('I','L')

# write modified dataframe to new txt file
peaksdb330.to_csv("/home/millieginty/Documents/git-repos/rot-mayer/data/processed/PeaksDB/TW_330_T2_undigested_combine_PTMopt_DB_FDR1.csv")

# check out the results
peaksdb330.head()
```

## NAAF correction and exporting files with AA and PTM totals:

```python
# made a new dataframe that contains the sums of certain columns in the modified
# peptide dataframe above 

index = ['sample total']

data = {'A': peaksdb330['A'].sum(),
        'C': peaksdb330['C'].sum(),
        'D': peaksdb330['D'].sum(),
        'E': peaksdb330['E'].sum(),
        'F': peaksdb330['F'].sum(),
        'G': peaksdb330['G'].sum(),
        'H': peaksdb330['H'].sum(),
        'I': peaksdb330['I'].sum(),
        'K': peaksdb330['K'].sum(),
        'L': peaksdb330['L'].sum(),
        'M': peaksdb330['M'].sum(),
        'N': peaksdb330['N'].sum(),
        'P': peaksdb330['P'].sum(),
        'Q': peaksdb330['Q'].sum(),
        'R': peaksdb330['R'].sum(),
        'S': peaksdb330['S'].sum(),
        'T': peaksdb330['T'].sum(),
        'V': peaksdb330['V'].sum(),
        'W': peaksdb330['W'].sum(),
        'Y': peaksdb330['Y'].sum(),
        'c-carb': peaksdb330['c-carb'].sum(),
        'm-oxid': peaksdb330['m-oxid'].sum(),
        'k-oxid': peaksdb330['k-oxid'].sum(),
        'p-oxid': peaksdb330['p-oxid'].sum(),
        'r-oxid': peaksdb330['r-oxid'].sum(),
        'y-oxid': peaksdb330['y-oxid'].sum(),
        'n-deam': peaksdb330['n-deam'].sum(),
        'k-meth': peaksdb330['k-meth'].sum(),
        'r-meth': peaksdb330['r-meth'].sum(),
        'q-pyro': peaksdb330['q-pyro'].sum(),
        'k-acet': peaksdb330['k-acet'].sum(),
        'Total area': peaksdb330['Area'].sum(),
        'Total length': peaksdb330['stripped length'].sum()
       }

totalpeaksdb330 = pd.DataFrame(data, columns=['A', 'C', 'D', 'E', 'F', 'G', 'H', 'I', 'K', 'L', \
                                              'M', 'N', 'P', 'Q', 'R', 'S', 'T', 'V', 'W', 'Y', \
                                              'c-carb', 'm-oxid', 'k-oxid', 'p-oxid', 'r-oxid', \
                                              'y-oxid', 'n-deam', 'k-meth', 'r-meth', 'q-pyro', \
                                              'k-acet', 'Total area', 'Total length'], index=index)

# calculate percentage of C's with carb (should be 1.0)
totalpeaksdb330['% C w/ carb'] = totalpeaksdb330['c-carb'] / totalpeaksdb330['C'] 

# calculate percentage of M's that are oxidized
totalpeaksdb330['% M w/ oxid'] = totalpeaksdb330['m-oxid'] / totalpeaksdb330['M'] 

# calculate percentage of K's that are oxidized
totalpeaksdb330['% K w/ oxid'] = totalpeaksdb330['k-oxid'] / totalpeaksdb330['K'] 

# calculate percentage of P's that are oxidized
totalpeaksdb330['% P w/ oxid'] = totalpeaksdb330['p-oxid'] / totalpeaksdb330['P'] 

# calculate percentage of R's that are oxidized
totalpeaksdb330['% R w/ oxid'] = totalpeaksdb330['p-oxid'] / totalpeaksdb330['R'] 

# calculate percentage of Y's that are oxidized
totalpeaksdb330['% Y w/ oxid'] = totalpeaksdb330['y-oxid'] / totalpeaksdb330['Y'] 

# calculate percentage of N's that are deamidated
totalpeaksdb330['% N w/ deam'] = totalpeaksdb330['n-deam'] / totalpeaksdb330['N'] 

# calculate percentage of K's that are methylated
totalpeaksdb330['% K w/ meth'] = totalpeaksdb330['k-meth'] / totalpeaksdb330['K'] 

# calculate percentage of R's that are methylated
totalpeaksdb330['% R w/ meth'] = totalpeaksdb330['r-meth'] / totalpeaksdb330['R'] 

# calculate percentage of Q's that are pyro glu'd
totalpeaksdb330['% Q w/ pyro'] = totalpeaksdb330['q-pyro'] / totalpeaksdb330['Q'] 

# calculate percentage of K's that are acetylation
totalpeaksdb330['% K w/ acet'] = totalpeaksdb330['k-acet'] / totalpeaksdb330['K'] 

# calculate NAAF denominator for all peptides in dataset i
totalpeaksdb330['NAAF denom.'] = totalpeaksdb330['Total area'] / totalpeaksdb330['Total length']

# write modified dataframe to new txt file
totalpeaksdb330.to_csv("/home/millieginty/Documents/git-repos/rot-mayer/data/processed/PeaksDB/TW_330_T2_undigested_combine_PTMopt_DB_FDR1_totals.csv")

totalpeaksdb330.head()
```

```python
# use the calculated NAAF factor (in totalpeaksdb330 dataframe, above) to caluclate the NAAF 
# NAAF: normalized normalized area abundance factor

NAAF20 = 541640.333788

# we'll use -10lgP > 20 as an approximate cutoff for the <1.0% FDR which is what we actually have
peaksdb330['NAAF factor'] = (peaksdb330['NAAF num.'])/NAAF20

# make a dataframe that contains only what we need: sequences, AAs, PTMs
peaksdb330_NAAF = peaksdb330[['stripped peptide', 'NAAF factor', 'A', 'C', 'D', 'E', 'F', 'G', 'H', \
                              'I', 'K', 'L', 'M', 'N', 'P', 'Q', 'R', 'S', 'T', 'V', 'W', 'Y', \
                              'c-carb', 'm-oxid', 'k-oxid', 'p-oxid', 'r-oxid', 'y-oxid', 'n-deam', \
                              'k-meth', 'r-meth', 'q-pyro', 'k-acet']].copy()

# multiply the NAAF20 factor by the AAs to normalize its abundance by peak area and peptide length

peaksdb330_NAAF['A-NAAF20'] = peaksdb330_NAAF['A'] * peaksdb330['NAAF factor']
peaksdb330_NAAF['C-NAAF20'] = peaksdb330_NAAF['C'] * peaksdb330['NAAF factor']
peaksdb330_NAAF['D-NAAF20'] = peaksdb330_NAAF['D'] * peaksdb330['NAAF factor']
peaksdb330_NAAF['E-NAAF20'] = peaksdb330_NAAF['E'] * peaksdb330['NAAF factor']
peaksdb330_NAAF['F-NAAF20'] = peaksdb330_NAAF['F'] * peaksdb330['NAAF factor']
peaksdb330_NAAF['G-NAAF20'] = peaksdb330_NAAF['G'] * peaksdb330['NAAF factor']
peaksdb330_NAAF['H-NAAF20'] = peaksdb330_NAAF['H'] * peaksdb330['NAAF factor']
peaksdb330_NAAF['I-NAAF20'] = peaksdb330_NAAF['I'] * peaksdb330['NAAF factor']
peaksdb330_NAAF['K-NAAF20'] = peaksdb330_NAAF['K'] * peaksdb330['NAAF factor']
peaksdb330_NAAF['L-NAAF20'] = peaksdb330_NAAF['L'] * peaksdb330['NAAF factor']
peaksdb330_NAAF['M-NAAF20'] = peaksdb330_NAAF['M'] * peaksdb330['NAAF factor']
peaksdb330_NAAF['N-NAAF20'] = peaksdb330_NAAF['N'] * peaksdb330['NAAF factor']
peaksdb330_NAAF['P-NAAF20'] = peaksdb330_NAAF['P'] * peaksdb330['NAAF factor']
peaksdb330_NAAF['Q-NAAF20'] = peaksdb330_NAAF['Q'] * peaksdb330['NAAF factor']
peaksdb330_NAAF['R-NAAF20'] = peaksdb330_NAAF['R'] * peaksdb330['NAAF factor']
peaksdb330_NAAF['S-NAAF20'] = peaksdb330_NAAF['S'] * peaksdb330['NAAF factor']
peaksdb330_NAAF['T-NAAF20'] = peaksdb330_NAAF['T'] * peaksdb330['NAAF factor']
peaksdb330_NAAF['V-NAAF20'] = peaksdb330_NAAF['V'] * peaksdb330['NAAF factor']
peaksdb330_NAAF['W-NAAF20'] = peaksdb330_NAAF['W'] * peaksdb330['NAAF factor']
peaksdb330_NAAF['Y-NAAF20'] = peaksdb330_NAAF['Y'] * peaksdb330['NAAF factor']

# multiply the NAAF20 factor by the PTMs normalize its abundance by peak area and peptide length

peaksdb330_NAAF['ccarb-NAAF20'] = peaksdb330_NAAF['c-carb'] * peaksdb330_NAAF['NAAF factor']
peaksdb330_NAAF['moxid-NAAF20'] = peaksdb330_NAAF['m-oxid'] * peaksdb330_NAAF['NAAF factor']
peaksdb330_NAAF['koxid-NAAF20'] = peaksdb330_NAAF['k-oxid'] * peaksdb330_NAAF['NAAF factor']
peaksdb330_NAAF['poxid-NAAF20'] = peaksdb330_NAAF['p-oxid'] * peaksdb330_NAAF['NAAF factor']
peaksdb330_NAAF['roxid-NAAF20'] = peaksdb330_NAAF['r-oxid'] * peaksdb330_NAAF['NAAF factor']
peaksdb330_NAAF['yoxid-NAAF20'] = peaksdb330_NAAF['y-oxid'] * peaksdb330_NAAF['NAAF factor']
peaksdb330_NAAF['ndeam-NAAF20'] = peaksdb330_NAAF['n-deam'] * peaksdb330_NAAF['NAAF factor']
peaksdb330_NAAF['kmeth-NAAF20'] = peaksdb330_NAAF['k-meth'] * peaksdb330_NAAF['NAAF factor']
peaksdb330_NAAF['rmeth-NAAF20'] = peaksdb330_NAAF['r-meth'] * peaksdb330_NAAF['NAAF factor']
peaksdb330_NAAF['qpyro-NAAF20'] = peaksdb330_NAAF['q-pyro'] * peaksdb330_NAAF['NAAF factor']
peaksdb330_NAAF['kacet-NAAF20'] = peaksdb330_NAAF['k-acet'] * peaksdb330_NAAF['NAAF factor']

# write the dataframe to a new csv
peaksdb330_NAAF.to_csv("/home/millieginty/Documents/git-repos/rot-mayer/data/processed/PeaksDB/TW_330_T2_undigested_combine_PTMopt_DB_FDR1_NAAF.csv")

peaksdb330_NAAF.head()
```

```python
# made a new dataframe that contains the sums of NAAF normalized AAs for peaksdb330 peaksdb results
# also contains the sums of the NAAF-corrected PTMs occurances for each affected residue

index = ['sample total']

data = {'NAAF': peaksdb330_NAAF['NAAF factor'].sum(),
        'A': peaksdb330_NAAF['A'].sum(),
        'C': peaksdb330_NAAF['C'].sum(),
        'D': peaksdb330_NAAF['D'].sum(),
        'E': peaksdb330_NAAF['E'].sum(),
        'F': peaksdb330_NAAF['F'].sum(),
        'G': peaksdb330_NAAF['G'].sum(),
        'H': peaksdb330_NAAF['H'].sum(),
        'I': peaksdb330_NAAF['I'].sum(),
        'K': peaksdb330_NAAF['K'].sum(),
        'L': peaksdb330_NAAF['L'].sum(),
        'M': peaksdb330_NAAF['M'].sum(),
        'N': peaksdb330_NAAF['N'].sum(),
        'P': peaksdb330_NAAF['P'].sum(),
        'Q': peaksdb330_NAAF['Q'].sum(),
        'R': peaksdb330_NAAF['R'].sum(),
        'S': peaksdb330_NAAF['S'].sum(),
        'T': peaksdb330_NAAF['T'].sum(),
        'V': peaksdb330_NAAF['V'].sum(),
        'W': peaksdb330_NAAF['W'].sum(),
        'Y': peaksdb330_NAAF['Y'].sum(),
        'c-carb': peaksdb330_NAAF['ccarb-NAAF20'].sum(),
        'm-oxid': peaksdb330_NAAF['moxid-NAAF20'].sum(),
        'k-oxid': peaksdb330_NAAF['koxid-NAAF20'].sum(),
        'p-oxid': peaksdb330_NAAF['poxid-NAAF20'].sum(),
        'r-oxid': peaksdb330_NAAF['roxid-NAAF20'].sum(),
        'y-oxid': peaksdb330_NAAF['yoxid-NAAF20'].sum(),
        'n-deam': peaksdb330_NAAF['ndeam-NAAF20'].sum(),
        'k-meth': peaksdb330_NAAF['kmeth-NAAF20'].sum(),
        'r-meth': peaksdb330_NAAF['rmeth-NAAF20'].sum(),
        'q-pyro': peaksdb330_NAAF['qpyro-NAAF20'].sum(),
        'k-acet': peaksdb330_NAAF['kacet-NAAF20'].sum()
       }

totalpeaksdb330_NAAF = pd.DataFrame(data, columns=['NAAF', 'A', 'C', 'D', 'E', 'F', \
                                                   'G', 'H', 'I','K', 'L', 'M', \
                                                   'N', 'P', 'Q', 'R', 'S', \
                                                   'T', 'V', 'W', 'Y', 'c-carb', \
                                                   'm-oxid', 'k-oxid', 'p-oxid', 'r-oxid', \
                                                   'y-oxid', 'n-deam', 'k-meth', 'r-meth', \
                                                   'q-pyro', \
                                                   'k-acet'], index=index)

# calculate the NAAF-corrected % modified C, M, N, Q, K, and Rs


# calculate NAAF-corrected percentage of C's with carb (should be 1.0)
totalpeaksdb330_NAAF['% C w/ carb.'] = totalpeaksdb330_NAAF['c-carb'] / totalpeaksdb330_NAAF['C'] 

# calculate NAAF-corrected percentage of M's that are oxidized
totalpeaksdb330_NAAF['% M w/ oxid'] = totalpeaksdb330_NAAF['m-oxid'] / totalpeaksdb330_NAAF['M'] 

# calculate NAAF-corrected percentage of K's that are oxidized
totalpeaksdb330_NAAF['% K w/ oxid'] = totalpeaksdb330_NAAF['k-oxid'] / totalpeaksdb330_NAAF['K'] 

# calculate NAAF-corrected percentage of P's that are oxidized
totalpeaksdb330_NAAF['% P w/ oxid'] = totalpeaksdb330_NAAF['p-oxid'] / totalpeaksdb330_NAAF['P'] 

# calculate NAAF-corrected percentage of R's that are oxidized
totalpeaksdb330_NAAF['% R w/ oxid'] = totalpeaksdb330_NAAF['r-oxid'] / totalpeaksdb330_NAAF['R'] 

# calculate NAAF-corrected percentage of Y's that are oxidized
totalpeaksdb330_NAAF['% Y w/ oxid'] = totalpeaksdb330_NAAF['y-oxid'] / totalpeaksdb330_NAAF['Y'] 

# calculate NAAF-corrected percentage of N's that are deamidated
totalpeaksdb330_NAAF['% N w/ deam'] = totalpeaksdb330_NAAF['n-deam'] / totalpeaksdb330_NAAF['N'] 

# calculate NAAF-corrected percentage of K's that are methylated
totalpeaksdb330_NAAF['% K w/ meth'] = totalpeaksdb330_NAAF['k-meth'] / totalpeaksdb330_NAAF['K'] 

# calculate NAAF-corrected percentage of R's that are methylated
totalpeaksdb330_NAAF['% R w/ meth'] = totalpeaksdb330_NAAF['r-meth'] / totalpeaksdb330_NAAF['R'] 

# calculate NAAF-corrected percentage of Q's that are pyro glu'd
totalpeaksdb330_NAAF['% Q w/ pyro'] = totalpeaksdb330_NAAF['q-pyro'] / totalpeaksdb330_NAAF['Q'] 

# calculate NAAF-corrected percentage of K's that are methylated
totalpeaksdb330_NAAF['% K w/ acet'] = totalpeaksdb330_NAAF['k-acet'] / totalpeaksdb330_NAAF['K'] 

# calculate NAAF summed numerator over denominator (in above cell) for all peptides in dataset i: a check
totalpeaksdb330_NAAF['NAAF check'] = totalpeaksdb330_NAAF['NAAF'] / 541640.333788

# write modified dataframe to new txt file, same name + totals
totalpeaksdb330_NAAF.to_csv("/home/millieginty/Documents/git-repos/rot-mayer/data/processed/PeaksDB/TW_330_T2_undigested_combine_PTMopt_DB_FDR1_NAAF_totals.csv")

totalpeaksdb330_NAAF.head()
```
