### Evaluating and visualizing gene ontology (GO) terms from _Thalassiosira pseudonana_ reference data

#### Beginning with: GO term abundances in spreadsheets derived from UniProt _Thaps_ ref. proteome
    - downloaded June 10, 2021
    - protein sequences digested in silico with trypsin
    - tryptic peptides analyzed in UniPept (downloaded all cellular component GOs)

#### Goal: heatmaps of GO terms 

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import regex as re
from functools import reduce
```

```python
# read the thaps proteome data into pandas dataframes
thaps = pd.read_csv("/home/millieginty/Documents/git-repos/rot-mayer/data/thaps-ref/thaps-cc-go-sorted.csv", index_col=0)

thaps.head()
```

```python
sns.set(font_scale=1.2)  

fig, ax = plt.subplots(figsize=(5,10))

ax = sns.heatmap(thaps, cmap="Oranges", annot=True, fmt='g', linecolor = 'grey', linewidths=.01, \
                square=True, cbar_kws={'label': '% of all bacterial GO terms'}, label='big')

ax.set_ylabel('')

# turn the axis label
for item in ax.get_yticklabels():
    item.set_rotation(0)
    

for item in ax.get_xticklabels():
    item.set_rotation(45)
```

### GO terms predicted from Dyherman et al., 2012 Thaps proteome (phosphorus replete)

#### Downloaded Supplementary Table 4 from Dyherman et al., 2012 (PLoS One), 'The Transcriptome and Proteome of the Diatom Thalassiosira pseudonana Reveal a Diverse Phosphorus Stress Response'.

```python
# downloaded Supplementary Table 4 and made a new file with just peptide sequences
# they have trypic heads and tails to trim and asterix in places to get rid of

#read the CSV into a datadrame we name 'comet' using the pandas read_csv function
thaps_d = pd.read_csv("/home/millieginty/Documents/git-repos/rot-mayer/data/thaps-ref/dyherman_2012_peptides.csv")

# create a column with 'stripped' peptide sequences using strip
thaps_d['stripped peptide'] = thaps_d['Peptide'].str[2:].str[:-2].str.replace(r"*","")

# save to a csv file in the same dir

# keep only stripped peptide column
pep = thaps_d[["stripped peptide"]]

# write altered dataframe to new txt file
# used header and index parameters to get rid of 'Peptide' header and the indexing

pep.to_csv("/home/millieginty/Documents/git-repos/rot-mayer/data/thaps-ref/dyherman_2012_stripped_peptides.txt", header=False, index=False)

thaps_d.head()
```
