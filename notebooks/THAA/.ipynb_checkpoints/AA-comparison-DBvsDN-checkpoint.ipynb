### Plotting relative AA results from DB and DN peptidopmics using python for _T. weissflogii_ rot experiment

Starting with:

    Amino acid composition files for Peak DB and de novo peptides


Goal:

    Compare relative AA from this approach to peptides from peptidomics
    Make figure showing AA compositions of THAA and peptides over rot timepoints (Days 0,2,5,12)
    
Using:

    - pandas
    - matplotlib
    - numpy

```python
# LIBRARIES
#import pandas library for working with tabular data
import os
os.getcwd()
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import kde
#import regular expresson (regex)
import re
#check pandas version
pd.__version__
```

## Comparison to peptide AA composition - TD

### Reading in the csv of digested peptide relative AAs as a dataframe
### Just PeaksDB peptides, so only algal signal.

```python
# Read in the peptide NAAF-corrected relative AAs of NATURALLY DIGESTED ("undigested") peptides
# Database only (just algal)
# These are not the THAA comparable files - they include C, N, Q, W

ND_DB_Relnaaf = pd.read_csv("/home/millieginty/Documents/git-repos/rot-mayer/analyses/AA-composition/undigested-DB-RelNAAF-AAs.csv", index_col=0)

# Rename the columns without NAAF for simplicity

ND_DB_Relnaaf.columns = ['sample', 'A_Rel', 'C_Rel', 'D_Rel', 'E_Rel', 'F_Rel', \
                             'G_Rel', 'H_Rel', \
                             'K_Rel', 'I/L_Rel', 'M_Rel', 'N_Rel', \
                             'P_Rel', 'Q_Rel', 'R_Rel', \
                             'S_Rel', 'T_Rel', 'V_Rel', 'W_Rel', 'Y_Rel']

ND_DB_Relnaaf.head()
```

```python
# Read in the peptide NAAF-corrected relative AAs of TRYPSIN DIGESTED ("digested") peptides
# Database only (just algal)
# These are not the THAA comparable files - they include C, N, Q, W

TD_DB_Relnaaf = pd.read_csv("/home/millieginty/Documents/git-repos/rot-mayer/analyses/AA-composition/digested-DB-RelNAAF-AAs.csv", index_col=0)

# Rename the columns without NAAF for simplicity

TD_DB_Relnaaf.columns = ['sample', 'A_Rel', 'C_Rel', 'D_Rel', 'E_Rel', 'F_Rel', \
                             'G_Rel', 'H_Rel', \
                             'K_Rel', 'I/L_Rel', 'M_Rel', 'N_Rel', \
                             'P_Rel', 'Q_Rel', 'R_Rel', \
                             'S_Rel', 'T_Rel', 'V_Rel', 'W_Rel', 'Y_Rel']

TD_DB_Relnaaf.head()
```

### De novo peptides

```python
# Read in the peptide NAAF-corrected relative AAs of NATURALLY DIGESTED ("undigested") peptides
# These are not the THAA comparable files - they include C, N, Q, W
# Just DN

ND_DN_Relnaaf = pd.read_csv("/home/millieginty/Documents/git-repos/rot-mayer/analyses/AA-composition/undigested-DN-RelNAAF-AAs.csv", index_col=0)

# Rename the columns without NAAF for simplicity

ND_DN_Relnaaf.columns = ['sample', 'A_Rel', 'C_Rel', 'D_Rel', 'E_Rel', 'F_Rel', \
                             'G_Rel', 'H_Rel', \
                             'K_Rel', 'I/L_Rel', 'M_Rel', 'N_Rel', \
                             'P_Rel', 'Q_Rel', 'R_Rel', \
                             'S_Rel', 'T_Rel', 'V_Rel', 'W_Rel', 'Y_Rel']

ND_DN_Relnaaf.head()
```

```python
# Read in the peptide NAAF-corrected relative AAs of TRYPSIN DIGESTED ("digested") peptides
# These are not the THAA comparable files - they include C, N, Q, W
# Just DN

TD_DN_Relnaaf = pd.read_csv("/home/millieginty/Documents/git-repos/rot-mayer/analyses/AA-composition/digested-DN-RelNAAF-AAs.csv", index_col=0)

# Rename the columns without NAAF for simplicity

TD_DN_Relnaaf.columns = ['sample', 'A_Rel', 'C_Rel', 'D_Rel', 'E_Rel', 'F_Rel', \
                             'G_Rel', 'H_Rel', \
                             'K_Rel', 'I/L_Rel', 'M_Rel', 'N_Rel', \
                             'P_Rel', 'Q_Rel', 'R_Rel', \
                             'S_Rel', 'T_Rel', 'V_Rel', 'W_Rel', 'Y_Rel']

TD_DN_Relnaaf.head()
```

## Plot database vs de novo AA comp on 1:1 line

```python
# Make scatter plots of trypsin digested AA compositions
# Database vs de novo
# If they agree, they'll fall on a 1:1 slope

import matplotlib.lines as mlines
import matplotlib.transforms as mtransforms

Ala_x = TD_DN_Relnaaf.iloc[0:, 1]
Ala_y = TD_DB_Relnaaf.iloc[0:, 1]

Cys_x = TD_DN_Relnaaf.iloc[0:, 2]
Cys_y = TD_DB_Relnaaf.iloc[0:, 2]

Asp_x = TD_DN_Relnaaf.iloc[0:, 3]
Asp_y = TD_DB_Relnaaf.iloc[0:, 3]

Glu_x = TD_DN_Relnaaf.iloc[0:, 4]
Glu_y = TD_DB_Relnaaf.iloc[0:, 4]

Phe_x = TD_DN_Relnaaf.iloc[0:, 5]
Phe_y = TD_DB_Relnaaf.iloc[0:, 5]

Gly_x = TD_DN_Relnaaf.iloc[0:, 6]
Gly_y = TD_DB_Relnaaf.iloc[0:, 6]

His_x = TD_DN_Relnaaf.iloc[0:, 7]
His_y = TD_DB_Relnaaf.iloc[0:, 7]

Lys_x = TD_DN_Relnaaf.iloc[0:, 8]
Lys_y = TD_DB_Relnaaf.iloc[0:, 8]

Leu_x = TD_DN_Relnaaf.iloc[0:, 9]
Leu_y = TD_DB_Relnaaf.iloc[0:, 9]

Met_x = TD_DN_Relnaaf.iloc[0:, 10]
Met_y = TD_DB_Relnaaf.iloc[0:, 10]

Asn_x = TD_DN_Relnaaf.iloc[0:, 11]
Asn_y = TD_DB_Relnaaf.iloc[0:, 11]

Pro_x = TD_DN_Relnaaf.iloc[0:, 12]
Pro_y = TD_DB_Relnaaf.iloc[0:, 12]

Gln_x = TD_DN_Relnaaf.iloc[0:, 13]
Gln_y = TD_DB_Relnaaf.iloc[0:, 13]

Arg_x = TD_DN_Relnaaf.iloc[0:, 14]
Arg_y = TD_DB_Relnaaf.iloc[0:, 14]

Ser_x = TD_DN_Relnaaf.iloc[0:, 15]
Ser_y = TD_DB_Relnaaf.iloc[0:, 15]

Thr_x = TD_DN_Relnaaf.iloc[0:, 16]
Thr_y = TD_DB_Relnaaf.iloc[0:, 16]

Val_x = TD_DN_Relnaaf.iloc[0:, 17]
Val_y = TD_DB_Relnaaf.iloc[0:, 17]

Trp_x = TD_DN_Relnaaf.iloc[0:, 18]
Trp_y = TD_DB_Relnaaf.iloc[0:, 18]

Tyr_x = TD_DN_Relnaaf.iloc[0:, 19]
Tyr_y = TD_DB_Relnaaf.iloc[0:, 19]

fig = plt.figure(figsize=(8,8))
ax1 = fig.add_subplot(111)

line = mlines.Line2D([0, 2], [0, 2], color='grey')
transform = ax1.transAxes
line.set_transform(transform)
ax1.add_line(line)

# Above 1:1 line: BLUES

# Arginine
ax1.scatter(Arg_x, Arg_y, s=100, c='#d0d1e6', marker="^", label='Arginine', edgecolors='black')

# Lysine
ax1.scatter(Lys_x, Lys_y, s=100, c='#3690c0', marker="o", label='Lysine', edgecolors='black')

# Methinoine
ax1.scatter(Met_x, Met_y, s=100, c='#0570b0', marker="P", label='Methionine', edgecolors='black')

# Threonine
ax1.scatter(Thr_x, Thr_y, s=100, c='#045a8d', marker="o", label='Threonine', edgecolors='black')

# Proline
ax1.scatter(Pro_x, Pro_y, s=100, c='#023858', marker="s", label='Proline', edgecolors='black')

# New ones: PURPLES

# Cysteine
ax1.scatter(Cys_x, Cys_y, s=100, c='#bcbddc', marker="s", label='Cysteine', edgecolors='black')

# Asparagine
ax1.scatter(Asn_x, Asn_y, s=100, c='#756bb1', marker="o", label='Asparagine', edgecolors='black')

# Glutamine
ax1.scatter(Cys_x, Cys_y, s=100, c='#dadaeb', marker="P", label='Cysteine', edgecolors='black')

# Tryptophan
ax1.scatter(Trp_x, Trp_y, s=100, c='#4a1486', marker="^", label='Asparagine', edgecolors='black')

# Across 1:1 line: GREENS

# Valine
ax1.scatter(Val_x, Val_y, s=100, c='#31a354', marker="<", label='Valine', edgecolors='black')

# Tyrosine
ax1.scatter(Tyr_x, Tyr_y, s=120, c='#c7e9c0', marker=">", label='Tyrosine', edgecolors='black')

# Serine
ax1.scatter(Ser_x, Ser_y, s=100, c='#a1d99b', marker="s", label='Serine', edgecolors='black')

# Leucine/isoleucine
ax1.scatter(Leu_x, Leu_y, s=100, c='#41ab5d', marker="P", label='Leucine/isoleucine', edgecolors='black')

# Below 1:1 line: REDS

# Glutamic acid
ax1.scatter(Glu_x, Glu_y, s=100, c='#dd1c77', marker="^", label='Glutamic acid', edgecolors='black')

# Histidine
ax1.scatter(His_x, His_y, s=100, c='#fa9fb5', marker="o", label='Histidine', edgecolors='black')

# Phenylalanine
ax1.scatter(Phe_x, Phe_y, s=100, c='#fde0dd', marker="^", label='Phenylalanine', edgecolors='black')

# Alanine
ax1.scatter(Ala_x, Ala_y, s=100, c='#dd3497', marker="s", label='Alanine', edgecolors='black')

# Glycine
ax1.scatter(Gly_x, Gly_y, s=100, c='#ae017e', marker="P", label='Glycine', edgecolors='black')

# Aspartic acid
ax1.scatter(Asp_x, Asp_y, s=120, c='#7a0177', marker=">", label='Aspartic acid', edgecolors='black')

ax1.set_ylim(0,.23)
ax1.set_xlim(0,.23)

ax1.set_ylabel('TD datbase peptide mole fraction', labelpad = 25, fontsize = 14)
ax1.set_xlabel('TD de novo peptide mole fraction', labelpad = 25, fontsize = 14)
ax1.set_facecolor("#f7f4f9")
plt.legend(bbox_to_anchor=(1,1), loc="upper left")
plt.show()
```
