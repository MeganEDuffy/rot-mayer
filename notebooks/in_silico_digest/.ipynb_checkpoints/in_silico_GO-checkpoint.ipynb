### Evaluating and visualizing gene ontology (GO) terms associated with algal and bacterial proteins during degradation experiment

#### Beginning with: GO term abundances from UniPept LCA of tryptic peptides generated from an in silico digestion of identified algal proteins (many of which were found with non tryptic searches, so UniPept not feasible).

#### Goal: heatmaps of algal and bacterial GO terms over the 4 timepoints

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
from functools import reduce
```

```python
# read the data into pandas dataframes
D0_TD = pd.read_csv("/home/millieginty/Documents/git-repos/rot-mayer/analyses/unipept/GO-terms/trypsin-dig-proteinIDs/322-DB-protein-diatom-cc.csv")

# concatenate GO-term and Name columns
D0_TD['GO term'] = D0_TD['GO-term'] + " " + D0_TD['Name']
del D0_TD['GO-term']
del D0_TD['Name']

print('322:', len(D0_TD))

# read the data into pandas dataframes
D2_TD = pd.read_csv("/home/millieginty/Documents/git-repos/rot-mayer/analyses/unipept/GO-terms/trypsin-dig-proteinIDs/323-DB-protein-diatom-cc.csv")

# concatenate GO-term and Name columns
D2_TD['GO term'] = D2_TD['GO-term'] + " " + D2_TD['Name']
del D2_TD['GO-term']
del D2_TD['Name']

print('323:', len(D2_TD))

# read the data into pandas dataframes
D5_TD = pd.read_csv("/home/millieginty/Documents/git-repos/rot-mayer/analyses/unipept/GO-terms/trypsin-dig-proteinIDs/324-DB-protein-diatom-cc.csv")

# concatenate GO-term and Name columns
D5_TD['GO term'] = D5_TD['GO-term'] + " " + D5_TD['Name']
del D5_TD['GO-term']
del D5_TD['Name']

print('324:', len(D5_TD))

# read the data into pandas dataframes
D12_TD = pd.read_csv("/home/millieginty/Documents/git-repos/rot-mayer/analyses/unipept/GO-terms/trypsin-dig-proteinIDs/325-DB-protein-diatom-cc.csv")

# concatenate GO-term and Name columns
D12_TD['GO term'] = D12_TD['GO-term'] + " " + D12_TD['Name']
del D12_TD['GO-term']
del D12_TD['Name']

print('325:', len(D12_TD))
```

```python
# read the data into pandas dataframes
D0_ND = pd.read_csv("/home/millieginty/Documents/git-repos/rot-mayer/analyses/unipept/GO-terms/trypsin-dig-proteinIDs/329-DB-protein-diatom-cc.csv")

# concatenate GO-term and Name columns
D0_ND['GO term'] = D0_ND['GO-term'] + " " + D0_ND['Name']
del D0_ND['GO-term']
del D0_ND['Name']

print('329:', len(D0_ND))

# read the data into pandas dataframes
D2_ND = pd.read_csv("/home/millieginty/Documents/git-repos/rot-mayer/analyses/unipept/GO-terms/trypsin-dig-proteinIDs/330-DB-protein-diatom-cc.csv")

# concatenate GO-term and Name columns
D2_ND['GO term'] = D2_ND['GO-term'] + " " + D2_ND['Name']
del D2_ND['GO-term']
del D2_ND['Name']

print('330:', len(D2_ND))

# read the data into pandas dataframes
D5_ND = pd.read_csv("/home/millieginty/Documents/git-repos/rot-mayer/analyses/unipept/GO-terms/trypsin-dig-proteinIDs/331-DB-protein-diatom-cc.csv")

# concatenate GO-term and Name columns
D5_ND['GO term'] = D5_ND['GO-term'] + " " + D5_ND['Name']
del D5_ND['GO-term']
del D5_ND['Name']

print('331:', len(D5_ND))

# read the data into pandas dataframes
D12_ND = pd.read_csv("/home/millieginty/Documents/git-repos/rot-mayer/analyses/unipept/GO-terms/trypsin-dig-proteinIDs/332-DB-protein-diatom-cc.csv")

# concatenate GO-term and Name columns
D12_ND['GO term'] = D12_ND['GO-term'] + " " + D12_ND['Name']
del D12_ND['GO-term']
del D12_ND['Name']

print('332:', len(D12_ND))
```

```python
# join the dataframes
d0d2 = pd.merge(D0_TD, D2_TD, on = ['GO term'], how = 'outer')

print('combined rows:', len(d0d2))

# rename columns
names = ['Day 0 TD', 'GO term', 'Day 2 TD']
         
d0d2.columns = names
```

```python
# join the dataframes
d0d2d5 = pd.merge(d0d2, D5_TD, on = ['GO term'], how = 'outer')

print('combined rows:', len(d0d2d5))

# rename columns
names = ['Day 0 TD', 'GO term', 'Day 2 TD', 'Day 5 TD']
         
d0d2d5.columns = names
```

```python
# join the dataframes
d0d2d5d12 = pd.merge(d0d2d5, D12_TD, on = ['GO term'], how = 'outer')

print('combined rows:', len(d0d2d5d12))

# rename columns
names = ['Day 0 TD', 'GO term', 'Day 2 TD', 'Day 5 TD', 'Day 12 TD']
         
d0d2d5d12.columns = names
```

```python
# join the dataframes
d0d2d5d12d0 = pd.merge(d0d2d5d12, D0_ND, on = ['GO term'], how = 'outer')

print('combined rows:', len(d0d2d5d12d0))

# rename columns
names = ['Day 0 TD', 'GO term', 'Day 2 TD', 'Day 5 TD', 'Day 12 TD', 'Day 0 ND']
         
d0d2d5d12d0.columns = names
```

```python
# join the dataframes
d0d2d5d12d0d2 = pd.merge(d0d2d5d12d0, D2_ND, on = ['GO term'], how = 'outer')

print('combined rows:', len(d0d2d5d12d0d2))

# rename columns
names = ['Day 0 TD', 'GO term', 'Day 2 TD', 'Day 5 TD', 'Day 12 TD', 'Day 0 ND', 'Day 2 ND']
         
d0d2d5d12d0d2.columns = names
```

```python
# join the dataframes
d0d2d5d12d0d2d5 = pd.merge(d0d2d5d12d0d2, D5_ND, on = ['GO term'], how = 'outer')

print('combined rows:', len(d0d2d5d12d0d2d5))

# rename columns
names = ['Day 0 TD', 'GO term', 'Day 2 TD', 'Day 5 TD', 'Day 12 TD', 'Day 0 ND', 'Day 2 ND', 'Day 5 ND']
         
d0d2d5d12d0d2d5.columns = names
```

```python
# join the dataframes
d0d2d5d12d0d2d5d12 = pd.merge(d0d2d5d12d0d2d5, D12_ND, on = ['GO term'], how = 'outer')

print('combined rows:', len(d0d2d5d12d0d2d5d12))

# rename columns
names = ['Day 0 TD', 'GO term', 'Day 2 TD', 'Day 5 TD', 'Day 12 TD', \
         'Day 0 ND', 'Day 2 ND', 'Day 5 ND', 'Day 12 ND']
         
d0d2d5d12d0d2d5d12.columns = names

d0d2d5d12d0d2d5d12.set_index('GO term', inplace = True)

combineGO = d0d2d5d12d0d2d5d12

# write combined GO term dataframe to a csv
combineGO.to_csv("/home/millieginty/Documents/git-repos/rot-mayer/analyses/unipept/GO-terms/GO-CC-in_silico_combined.csv")

combineGO.head()
```

```python
# plot heatmap

fig, ax = plt.subplots(figsize=(15,20))

ax = sns.heatmap(combineGO, cmap="Greens", annot=True, linecolor = 'grey', linewidths=.01, \
                square=False, cbar_kws={'label': '# of peptides'})

ax.set_ylabel('')

# turn the axis label
for item in ax.get_yticklabels():
    item.set_rotation(0)

for item in ax.get_xticklabels():
    item.set_rotation(45)
```

```python
# rearranged and condensed the csv manually in Libre Calc
# read in new csv:

in_sil = pd.read_csv("/home/millieginty/Documents/git-repos/rot-mayer/analyses/unipept/GO-terms/GO-CC-in_silico_combined_sorted.csv", index_col=0)

# also made csvs separating trypsin and naturally digested peptide GO terms

in_sil_td = pd.read_csv("/home/millieginty/Documents/git-repos/rot-mayer/analyses/unipept/GO-terms/GO-CC-in_silico_combined_sorted_td.csv", index_col=0)
in_sil_nd = pd.read_csv("/home/millieginty/Documents/git-repos/rot-mayer/analyses/unipept/GO-terms/GO-CC-in_silico_combined_sorted_nd.csv", index_col=0)

in_sil.head()
```

```python
sns.set(font_scale=1.1)  

fig, ax = plt.subplots(figsize=(10,10))

ax = sns.heatmap(in_sil, cmap="Greens", annot=True, fmt='g', linecolor = 'grey', linewidths=.01, \
                square=False, cbar_kws={'label': '# of peptides'}, label='big')

ax.set_ylabel('')

# turn the axis label
for item in ax.get_yticklabels():
    item.set_rotation(0)
    

for item in ax.get_xticklabels():
    item.set_rotation(45)
```

```python
# plot heatmap with treatments separated

sns.set(font_scale=1.2)  

f,(ax1,ax2) = plt.subplots(1,2,sharey=True, figsize=(8, 10))

plt.subplots_adjust(left=None, bottom=None, right=None, top=None, wspace=0.1, hspace=0.01)


labels = ['Day 0', 'Day 2', 'Day 5', 'Day 12']

g1 = sns.heatmap(in_sil_td, cbar=False, ax=ax1, cmap="Greens", annot=True, fmt='g', linecolor = 'grey', linewidths=.01, \
                square=False, label='big')

g1.set_ylabel('')
g1.set_xticklabels(labels, rotation=40)
g1.set_xlabel('Trypsin digested', fontsize = 14)


g2 = sns.heatmap(in_sil_nd, ax=ax2, cmap="Greens", annot=True, fmt='g', linecolor = 'grey', linewidths=.01, \
                square=False, cbar_kws={'label': '# of peptides'}, label='big')

g2.set_ylabel('')
g2.set_xticklabels(labels, rotation=40)
g2.set_xlabel('Naturally digested', fontsize = 14)
```

```python
# plot heatmap with treatments separated

import textwrap

sns.set(font_scale=1.2)  

f,(ax1,ax2,ax3) = plt.subplots(1,3, 
            gridspec_kw={'width_ratios':[1,1.25,1]}, figsize = (10,8))

ylabels = ['Chloroplast', 'Cell membrane', 'Mitochondria', 'Endoplasmic r.', 'Golgi', 'Ribosome', \
          'Nucleus', 'Cytoplasm']

plt.tight_layout(pad= 0.1)

ax1.get_shared_y_axes().join(ax2,ax3)

plt.subplots_adjust(left=None, bottom=None, right=None, top=None, wspace=0.1, hspace=0.01)

plt.subplots_adjust(left=0.15, bottom=0.2)


labels = ['Day 0', 'Day 2', 'Day 5', 'Day 12']
ylabels = ['', '', '', '', '', '', '', '']
dlabels = ['']

g1 = sns.heatmap(in_sil_td, cbar=False, ax=ax1, cmap="Greens", annot=True, fmt='g', linecolor = 'grey', linewidths=.01, \
                square=False, label='big')

g1.set_ylabel('')
g1.set_xticklabels(labels, rotation=40)
g1.set_xlabel('Trypsin digested', fontsize = 16)


g2 = sns.heatmap(in_sil_nd, ax=ax2, cmap="Greens", annot=True, fmt='g', linecolor = 'grey', linewidths=.01, \
                square=False, cbar_kws={'label': '# of peptides'}, label='big')

g2.set_ylabel('')
g2.set_yticklabels(ylabels)
g2.set_xticklabels(labels, rotation=40)
g2.set_xlabel('Naturally digested', fontsize = 16)


g3 = sns.heatmap(dyr_t, cmap="Oranges", annot=True, fmt='g', linecolor = 'grey', linewidths=.01, \
                square=True, cbar_kws={'label': '# of peptides'}, label='big')

g3.set_ylabel('')
g3.set_yticklabels(ylabels)
g3.set_xlabel(textwrap.fill('Thalassiosira pseudonana proteomes', 13), fontsize = 16)
g3.set_xticklabels(dlabels)

plt.savefig('/home/millieginty/Documents/git-repos/rot-mayer/figures/manuscript-files/fig2-GO.png', bbox_inches='tight', dpi=400)
```
