### Comparing peptide overlap of trypsin digested and naturally digested fractions

#### Can't comapre sequences by perfect overlap, since they're tryptic and non-tryptic, respectively. 

#### Instead, in our intial processesing we converted the PTM-stripped, tryptic peptidse in each timepoint to a FASTA. We will align with the non-tryptic peptides using PepExplorer - which will equate leucine and isoleucine. 

#### Then, compared aligned non-tryptic peptides to total non-tryptic pool using VennDrawer and saved results in `analyses/pepexplorer/Trypsin-vs-ND` as a text file. Then we can recerate the Venn diagrams above but with a more accurate overlap between peptides that accounts for the different digestion strategies.

### Couldn't figure it out in python, but [Galaxy](usegalaxy.org/)'s `Filter FASTA on the headers` tool work's perfectly, so I output the results from that in the VennDrawer output analyses subdirectory. 

### The process:

   1. Take the overlapping and ND-only peptides from VennDrawer output in `rot-mayer/analyses/VennDrawer/Trypsin-vs-DN` and make a text file'.
     
   2. Load into Galaxy along with the MMETSP (Marine Microbial Eukaryotic Transcriptome Sequencing Project) and UniProt _T. weiss._ protein database in `rot-mayer/data/database-fastas` in single line format (see above).
     
   3. Use the `Filter FASTA on headers and/or sequences` tool to generate a fasta of just the overlapping IDs.
     
   4. Save the new fasta in `analyses/VennDrawer/Trypsin-vs-DN` sample subdirectory.
   5. Perform an in silico trypsin digest on the shared protein fastas using Protein Digestion Simulator.
   6. Input the tryptic peptides into UniPept for functional analysis with GO term outputs.
   7. Save GO term outputs in `analyses/VennDrawer/Trypsin-vs-DN` sample subdirectory.

### Merge all the ND (naturally digested only) protein GO terms into a dataframe to make a heatmap:

```python
# read the data into pandas dataframes
D0_ND = pd.read_csv("/home/millieginty/Documents/git-repos/rot-mayer/analyses/VennDrawer/Trypsin-vs-DN/T0-protein-overlap/T0-ND-only/T0-329_only_protein_peptide_diatom_GO.csv")

# concatenate GO-term and Name columns
D0_ND['GO term'] = D0_ND['GO-term'] + " " + D0_ND['Name']
del D0_ND['GO-term']
del D0_ND['Name']

print('322:', len(D0_ND))

# read the data into pandas dataframes
D2_ND = pd.read_csv("/home/millieginty/Documents/git-repos/rot-mayer/analyses/VennDrawer/Trypsin-vs-DN/T2-protein-overlap/T2-ND-only/T2-330_only_protein_peptide_diatom_GO.csv")

# concatenate GO-term and Name columns
D2_ND['GO term'] = D2_ND['GO-term'] + " " + D2_ND['Name']
del D2_ND['GO-term']
del D2_ND['Name']

print('323:', len(D2_ND))

# read the data into pandas dataframes
D5_ND = pd.read_csv("/home/millieginty/Documents/git-repos/rot-mayer/analyses/VennDrawer/Trypsin-vs-DN/T5-protein-overlap/T5-ND-only/T5-331_only_protein_peptide_diatom_GO.csv")

# concatenate GO-term and Name columns
D5_ND['GO term'] = D5_ND['GO-term'] + " " + D5_ND['Name']
del D5_ND['GO-term']
del D5_ND['Name']

print('324:', len(D5_ND))

# read the data into pandas dataframes
D12_ND = pd.read_csv("/home/millieginty/Documents/git-repos/rot-mayer/analyses/VennDrawer/Trypsin-vs-DN/T12-protein-overlap/T12-ND-only/T12-332_only_protein_peptide_diatom_GO.csv")

# concatenate GO-term and Name columns
D12_ND['GO term'] = D12_ND['GO-term'] + " " + D12_ND['Name']
del D12_ND['GO-term']
del D12_ND['Name']

print('325:', len(D12_ND))
```

```python
# join the dataframes
d0d2 = pd.merge(D0_ND, D2_ND, on = ['GO term'], how = 'outer')

print('combined rows:', len(d0d2))

# rename columns
names = ['Day 0 ND', 'GO term', 'Day 2 ND']
         
d0d2.columns = names
```

```python
# join the dataframes
d0d2d5 = pd.merge(d0d2, D5_ND, on = ['GO term'], how = 'outer')

print('combined rows:', len(d0d2d5))

# rename columns
names = ['Day 0 ND', 'GO term', 'Day 2 ND', 'Day 5 ND']
         
d0d2d5.columns = names
```

```python
# join the dataframes
d0d2d5d12 = pd.merge(d0d2d5, D12_ND, on = ['GO term'], how = 'outer')

print('combined rows:', len(d0d2d5d12))

# rename columns
names = ['Day 0 ND', 'GO term', 'Day 2 ND', 'Day 5 ND', 'Day 12 ND']
         
d0d2d5d12.columns = names

d0d2d5d12.set_index('GO term', inplace = True)

combineGO = d0d2d5d12

# write combined GO term dataframe to a csv
combineGO.to_csv("/home/millieginty/Documents/git-repos/rot-mayer/analyses/VennDrawer/Trypsin-vs-DN/GO-CC-NDonly-in_silico_combined.csv")

combineGO.head()
```
