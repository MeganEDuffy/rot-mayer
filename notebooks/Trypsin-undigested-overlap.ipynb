### Comparing trypsin digested and non digested  peptides for each _T. weisflogii_ rot sample:

Starting with:

    Peaks de novo results of PTM-optimized sequencing, with leucine and isoleuce equated (DN)
    PeaksDB de novo-assisted results from PTM-optimized searches, with leucine and isoleuce equated (DB)

Goal:

    Venn diagrams of de novo and database peptide for each timepoint
    Venn diagrams of proteins predicted from each peptide set at each timepoint
    Venn diagram of peptide GO terms (cellular compartment, to start)
    
Using:

    - pandas
    - matplotlib
    - numpy
    - venn2

```python
# LIBRARIES
#import pandas library for working with tabular data
import os
os.getcwd()
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import kde
#import regular expresson (regex)
import re
#from matplotlib_venn import venn2
#from matplotlib_venn import venn3
from matplotlib_venn import venn2, venn2_circles
#check pandas version
pd.__version__
```

```python
# Day 0 database peptides

plt.figure(figsize=(4,5))
v = venn2(subsets=(710, 176, 67), \
          set_labels = ('Trypsin-digested DB', 'Non-digested DB'))
v.get_patch_by_id('100').set_alpha(1.0)

v.get_patch_by_id('100').set_color('#e6550d')

v.get_patch_by_id('100').set_lw(1.5)
v.get_patch_by_id('010').set_lw(1.5)


plt.tight_layout()

# save figure in analysis and also in manuscript git repos
#plt.savefig("/home/millieginty/Documents/git-repos/2017-etnp/data/pro2020/RAL4_95_MED2_trypsin/TPP/MED4-Comet_ProtProp-overlap.png", dpi=300, transparent=True)
#plt.savefig("/home/millieginty/Documents/git-repos/denovo-etnp/MED4-Comet_ProtProp-overlap.png", dpi=300, transparent=True)
plt.show()
```
