```python
import numpy as np 
import pandas as pd

#create dataset

day = ['Day 0', 'Day 2', 'Day 5', 'Day 12']

algal_cells = [4.01E+09, 3.09E+09, 3.25E+09, 1.70E+09]
bacterial_cells = [2.95E+10, 6.43E+10, 1.23E+11, 1.33E+11]

chl = [2282.14, 1156.52, 141.79, 101.73]
pheo = [18.26, 716.19, 1976.7, 1793.13]

df = pd.DataFrame({'day': day,
                   'algal_cells': algal_cells, 
                   'bacterial_cells': bacterial_cells,
                   'chl': chl,
                   'pheo': pheo})

df.set_index('day')

df.head(4)
```

```python
#plot individual lines

fig, ax = plt.subplots(2, 2, sharex = True)

ax[0][0].plot(df['algal_cells'], color = '#a1d99b', label='Algal cells')
ax[0, 0].set_title('b) Algal cells', pad=13)
ax[0, 0].set_ylabel('cell count')

ax[0][1].plot(df['bacterial_cells'], color = '#feb24c', label='Bacterial cells')
ax[0, 1].set_title('c) Bacterial cells', pad=13)
ax[0, 1].set_ylabel('cell count')

ax[1][0].plot(df['chl'], color = '#a1d99b', linestyle='dashed', label='Chlorophyll')
ax[1, 0].set_title('d) chlorophyll', pad=13)
ax[1, 0].set_ylabel('ug/L')

ax[1][1].plot(df['pheo'], color = '#636363', linestyle='dashed', label='Pheophytin')
ax[1, 1].set_title('e) pheophytin', pad=13)
ax[1, 1].set_ylabel('ug/L')
ax[1, 1].set_xticklabels(df['day'])

# set the spacing between subplots
plt.subplots_adjust(left=0.1,
                    bottom=0.1, 
                    right=0.9, 
                    top=0.9, 
                    wspace=0.4, 
                    hspace=0.5)

fig.text(0.5, -0.03, 'Rot progression', size = 14, ha='center')

#display plot
#plt.legend()
plt.show()
```
