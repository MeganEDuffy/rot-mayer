### Evaluating and visualizing NAAF-adjusted bacterial peptide composition during rot experiment

#### Beginning with: bacterial LCA assigments adjusted by NAAF factor in csv, manually created from bacterial LCA NAAF csvs

#### Goal: heatmaps and stacked bar plot bacterial peptide contribution over the 4 timepoints

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
from functools import reduce
```

```python
# rearranged and condensed the csv manually in Libre Calc
# all peptide counts adjusted by NAAF numerators and sample denominator
# read in new csv:

bac_NAAF = pd.read_csv("/home/millieginty/Documents/git-repos/rot-mayer/analyses/unipept/lca/bacterial-comp-NAAF/Rot-bacterial-comp_DN_NAAF_sorted.csv", index_col=0)

bac_NAAF.head()
```

```python
sns.set(font_scale=1.1)  

fig, ax = plt.subplots(figsize=(8,10))

ax = sns.heatmap(bac_NAAF, cmap="BuPu", annot=True, fmt='f', linecolor = 'grey', linewidths=.01, \
                square=False, cbar_kws={'label': '# of peptides'}, label='big')

ax.set_ylabel('')

# turn the axis label
for item in ax.get_yticklabels():
    item.set_rotation(0)
    

for item in ax.get_xticklabels():
    item.set_rotation(45)
```

```python
bac_TNAAF = bac_NAAF.transpose()

ax = bac_TNAAF.plot(kind='bar',stacked=True, figsize = (5,7))

ax.set_xlabel("Rot progression", fontsize = 16)

ax.set_ylabel("NAAF-adjusted peptide abundance", labelpad=10, fontsize = 16)

labels = ['Day 0', 'Day 2', 'Day 5', 'Day 12']

ax.set_xticklabels(labels, rotation = 45, size = 14)
plt.yticks(fontsize=14)

plt.legend(bbox_to_anchor=(1,1), loc="upper left")
```

```python
# rearranged and condensed the csv manually in Libre Calc
# PERCENT of all NAAF terms
# read in new csv:

bac_PER = pd.read_csv("/home/millieginty/Documents/git-repos/rot-mayer/analyses/unipept/lca/bacterial-comp-NAAF/Rot-bacterial-comp_DN_percent_sorted.csv", index_col=0)

bac_PER.head()
```

```python
sns.set(font_scale=1.1)  

fig, ax = plt.subplots(figsize=(8,10))

ax = sns.heatmap(bac_PER, cmap="Greens", annot=True, fmt='f', linecolor = 'grey', linewidths=.01, \
                square=False, cbar_kws={'label': '# of peptides'}, label='big')

ax.set_ylabel('')

# turn the axis label
for item in ax.get_yticklabels():
    item.set_rotation(0)
    

for item in ax.get_xticklabels():
    item.set_rotation(45)
```

```python
bac_TPER = bac_PER.transpose()

ax = bac_TPER.plot(kind='bar',stacked=True, figsize = (5,7))

ax.set_xlabel("Rot progression", fontsize = 16)

ax.set_ylabel("Relative peptide abundance", labelpad=10, fontsize = 16)

labels = ['Day 0', 'Day 2', 'Day 5', 'Day 12']

ax.set_xticklabels(labels, rotation = 45, size = 14)
plt.yticks(fontsize=14)

plt.legend(bbox_to_anchor=(1,1), loc="upper left")
```
