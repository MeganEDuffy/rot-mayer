```python
import matplotlib.pyplot as plt

# Some data
labels = 'Diatoms', 'Bacteria'
labels_no = '', ''
colors = '#a1d99b', '#fff7bc'

fracs_0 = [0.747589938151764, 0.252410061848236]
fracs_2 = [0.707973191695006, 0.292026808304994]
fracs_5 = [0.805886784961324, 0.194113215038676]
fracs_12 = [0.614692704542747, 0.385307295457253]

# Make figure and axes
fig, axs = plt.subplots(2, 2)

# Day 0
axs[0, 0].pie(fracs_0, labels=labels, colors = colors, \
              autopct='%1.1f%%', shadow=True, radius=1.5)
axs[0, 0].set_title('Day 0', pad=13)

# Day 2
axs[1, 0].pie(fracs_2, labels=labels_no, colors = colors, \
              autopct='%1.1f%%', shadow=True, radius=1.5)
axs[1, 0].set_title('Day 2', pad=13)

# Day 5
axs[0, 1].pie(fracs_5, labels=labels_no, colors = colors, \
              autopct='%1.1f%%', shadow=True, radius=1.5)
axs[0, 1].set_title('Day 5', pad=13)

# Day 12
axs[1, 1].pie(fracs_12, labels=labels_no, colors = colors, \
              autopct='%1.1f%%', shadow=True, radius=1.5)
axs[1, 1].set_title('Day 12', pad=13)

# set the spacing between subplots
plt.subplots_adjust(left=0.1,
                    bottom=0.1, 
                    right=0.9, 
                    top=0.9, 
                    wspace=0.3, 
                    hspace=0.5)



plt.show()
```

```python
import numpy as np 
import pandas as pd

#create dataset

day = ['Day 0', 'Day 2', 'Day 5', 'Day 12']

algal_cells = [4.01E+09, 3.09E+09, 3.25E+09, 1.70E+09]
bacterial_cells = [2.95E+10, 6.43E+10, 1.23E+11, 1.33E+11]

chl = [2282.14, 1156.52, 141.79, 101.73]
pheo = [18.26, 716.19, 1976.7, 1793.13]

carbon = [896.5, 765.6, 646.2, 619.6]
nitrogen = [96.2, 100.9, 100.3, 86.1]

ehaa_c = [179.88, 174.67, 139.71, 111.66]
ehaa_n = [59.96, 58.22, 46.57, 37.22]

nonehaa_c = [36.26, 42.63, 53.78, 48.87]
nonehaa_n = [716.65, 590.95, 506.47, 507.96]

df = pd.DataFrame({'day': day,
                   'algal_cells': algal_cells, 
                   'bacterial_cells': bacterial_cells,
                   'chl': chl,
                   'pheo': pheo,
                   'carbon': carbon,
                   'nitrogen': nitrogen,
                   'ehaa_c': ehaa_c,
                   'ehaa_n': ehaa_n,
                   'nonehaa_c': nonehaa_c,
                   'nonehaa_n': nonehaa_n})

df.set_index('day', inplace=True)

df.head(4)
```

```python
#plot individual lines

fig, ax = plt.subplots(2, 2, sharex = True, figsize=(8, 6))

# Algal cells:

ax[0][0].plot(df['algal_cells'], color = '#a1d99b', label='Algal cells')
ax[0][0].set_title('a) Algal cells', pad=13)
ax[0][0].set_ylabel('cell count')

# Bacterial cells:

ax[0][1].plot(df['bacterial_cells'], color = '#feb24c', label='Bacterial cells')
ax[0][1].set_title('b) Bacterial cells', pad=13)
ax[0][1].set_ylabel('cell count')

# Pigments

ax[1,0].plot(df['chl'], color = '#a1d99b', linestyle='dashed', label='chlorophyll')
ax[1,0].set_title('c) Pigments', pad=13)
ax[1,0].set_ylabel('ug/L')
ax[1,0].set_xticklabels(df.index, rotation = 30)
ax[1,0].legend(loc="upper right")
ax[1,0].plot(df['pheo'], color = '#636363', linestyle='dashed', label='pheophytin')
ax[1,0].legend(loc="upper right")

# Total C & N

ax[1,1].plot(df['carbon'], color = '#756bb1', linestyle='dashed', label='carbon')
ax[1,1].plot(df['nitrogen'], color = '#de2d26', label='nitrogen')
ax[1,1].set_title('d) POC & PON', pad=13)
ax[1,1].set_ylabel('mg/L')
ax[1,1].set_xticklabels(df.index, rotation = 30)
ax[1,1].legend(loc="upper right")

# EHAA C & N
# Decided to take out - June 2021

#ax[1,1].plot(df['ehaa_c'], color = '#756bb1', linestyle='dashed', label='EHAA C')
#ax[1,1].plot(df['ehaa_n'], color = '#de2d26', label='EHAA N')
#ax[1,1].set_title('e) EHAA C & N', pad=13)
#ax[1,1].set_ylabel('mg/L')
#ax[1,1].set_xticklabels(df.index, rotation = 30)
#ax[1,1].legend(loc="upper right")

# Total:EHAA C & N

#ax[1,2].plot(df['nonehaa_c'], color = '#756bb1', linestyle='dashed', label='non EHAA C')
#ax[1,2].plot(df['nonehaa_n'], color = '#de2d26', label='non EHAA N')
#ax[1,2].set_title('f) Non-EHAA C & N', pad=13)
#ax[1,2].set_ylabel('mg/L')
#ax[1,2].set_xticklabels(df.index, rotation = 30)
#ax[1,2].legend(loc="upper right")

# set the spacing between subplots
plt.subplots_adjust(left=0.1,
                    bottom=0.1, 
                    right=0.9, 
                    top=0.9, 
                    wspace=0.4, 
                    hspace=0.3)

#fig.text(0.5, -0.09, 'Rot progression', size = 14, ha='center')

#display plot
#plt.legend()
plt.show()
```
