### Sorting and visualizing UniPept lowest common ancestor analyses over diatom degradation experiment

#### Beginning with: LCA peptides in csvs

#### Goal: spectral abundance-corrected (NAAF) algal and bacterial peptide over the 4 timepoints

## Issue: the `join` commands at the end for diatom and bacterial peptides need to be run independently from one another. To do this, run the diatom command, restart the kernal, then run the bacterial. 

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
```

```python
# read the data into pandas dataframes
tryp_324_DN = pd.read_csv("/home/millieginty/Documents/git-repos/rot-mayer/analyses/unipept/lca/cleaned/T5-324-trypsin-lca-DN.csv")

tryp_324_DN.head(6)
```

```python
# keep only the rows where kingdom = Bacteria

tryp_324_DN_Bac1 = tryp_324_DN.loc[(tryp_324_DN['kingdom'] == 'Bacteria')]

tryp_324_DN_Bac2 = tryp_324_DN.loc[(tryp_324_DN['subkingdom'] == 'Bacteria')]

print('# of kingdom = ', len(tryp_324_DN_Bac1))

print('# of subkingdom = ', len(tryp_324_DN_Bac2))

frames = [tryp_324_DN_Bac1, tryp_324_DN_Bac2]
  
tryp_324_DN_Bac = pd.concat(frames)

print('# of total Bacteria peptides = ', len(tryp_324_DN_Bac))

tryp_324_DN_Bac.head()
```

```python
# keep only the rows where phylum = Bacillariophyta

tryp_324_DN_Dia1 = tryp_324_DN.loc[(tryp_324_DN['phylum'] == 'Bacillariophyta')]

tryp_324_DN_Dia2 = tryp_324_DN.loc[(tryp_324_DN['subphylum'] == 'Bacillariophyta')]

print('# of phylum = ', len(tryp_324_DN_Dia1))

print('# of subphylum = ', len(tryp_324_DN_Dia2))

frames = [tryp_324_DN_Dia1, tryp_324_DN_Dia2]
  
tryp_324_DN_Dia = pd.concat(frames)

print('# of total diatom peptides = ', len(tryp_324_DN_Dia))

print('colunm names:', tryp_324_DN_Dia.columns)

tryp_324_DN_Dia.head()
```

### Now I want to read in the file containing the stripped peptides with NAAF values
####  - NAAF stands for 'noramlized area abunace factor'

### I want to join the dataframes if they share an index (stripped peptide with equated leucine and isoleucines)
#### - That means I'll reindex the processed peptide file

```python
tryp_324_DN_NAAF = pd.read_csv("/home/millieginty/Documents/git-repos/rot-mayer/data/processed/I-L_NAAFs/T5_324_trypsin_DN_peptides.csv")

tryp_324_DN_NAAF.set_index('stripped_peptide')
tryp_324_DN_NAAF = tryp_324_DN_NAAF.loc[:, ~tryp_324_DN_NAAF.columns.str.contains('^Unnamed')]

tryp_324_DN_NAAF.rename(columns = {'stripped_peptide':'peptide'}, inplace = True)

print('# of total peptides = ', len(tryp_324_DN_NAAF))

print('column names:', tryp_324_DN_NAAF.columns)

tryp_324_DN_NAAF.head()
```

```python
# get the intersection of the diatom peptides and corresponding peptides w/ NAAFs

over = np.intersect1d(tryp_324_DN_Dia["peptide"], tryp_324_DN_NAAF["peptide"])

print(over)

#print(tryp_324_DN_Dia[tryp_324_DN_NAAF.peptide.isin(over)])

#print(tryp_324_DN_NAAF[tryp_324_DN_Dia.peptide.isin(over)])
```

```python
# join the dataframes if the peptide values are the same using 'join'
# since a couple are de novo only (more for bacteria), we won't have all the UniPept peptides overlap 

tryp_324_DN_Dia.set_index('peptide', inplace=True)
tryp_324_DN_NAAF.set_index('peptide', inplace=True)

tryp_324_DN_Diatoms = tryp_324_DN_Dia.join(tryp_324_DN_NAAF, how='left', rsuffix='_other')


print('# of total diatom peptides = ', len(tryp_324_DN_Diatoms))

tryp_324_DN_Diatoms.head()
```

```python
# write to a csv

tryp_324_DN_Diatoms.to_csv("/home/millieginty/Documents/git-repos/rot-mayer/data/processed/I-L_NAAFs/T5_324_trypsin_DN_NAAF_diatom_peptides.csv")
```

```python
# join the dataframes if the peptide values are the same using 'join'
# since a couple are de novo only (more for bacteria), we won't have all the UniPept peptides overlap 

tryp_324_DN_Bac.set_index('peptide', inplace=True)
tryp_324_DN_NAAF.set_index('peptide', inplace=True)

tryp_324_DN_Bacteria = tryp_324_DN_Bac.join(tryp_324_DN_NAAF, how='left', rsuffix='_other')


print('# of total Bacteria peptides = ', len(tryp_324_DN_Bacteria))

tryp_324_DN_Bacteria.head()
```

```python
# write to a csv

tryp_324_DN_Bacteria.to_csv("/home/millieginty/Documents/git-repos/rot-mayer/data/processed/I-L_NAAFs/T5_324_trypsin_DN_NAAF_bacteria_peptides.csv")
```
